```python
# ==========================================================
# rag_indexer.py ‚Äî Compatible LangChain 1.x
# ==========================================================
import os
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import FAISS

# Dossiers
PDF_DIR = "../data"
INDEX_DIR = "../embeddings/faiss_index"
os.makedirs(INDEX_DIR, exist_ok=True)

# Mod√®le d'embedding
MODEL = "sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2"

print("üìö Chargement des documents PDF‚Ä¶")
docs = []
for file in os.listdir(PDF_DIR):
    if file.endswith(".pdf"):
        loader = PyPDFLoader(os.path.join(PDF_DIR, file))
        docs.extend(loader.load())

print(f"‚úÖ {len(docs)} pages charg√©es.")

# D√©coupage en chunks
splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
texts = splitter.split_documents(docs)
print(f"‚úÇÔ∏è {len(texts)} chunks cr√©√©s.")

# Embeddings + Index
embeddings = HuggingFaceEmbeddings(model_name=MODEL)
print("üî¢ G√©n√©ration de l'index FAISS‚Ä¶")
db = FAISS.from_documents(texts, embeddings)
db.save_local(INDEX_DIR)
print(f"‚úÖ Index sauvegard√© dans : {INDEX_DIR}")
```
